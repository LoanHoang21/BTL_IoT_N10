# Model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.metrics import mean_squared_error, accuracy_score, mean_absolute_error, r2_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score

# Tách dữ liệu
X = df[['Year', 'Month', 'Day', 'Hour', 'Temperature (C)', 'Humidity (%)', 'Wind speed (km/h)', 'Pressure (hPa)', 'Light intensity (lux)']]
y_rain = df['Percent of rain (%)']
y_weather_type = df['Weather type']

# Chia tập dữ liệu
X_train, X_test, y_train_rain, y_test_rain = train_test_split(X, y_rain, test_size=0.2, random_state=42)
_,_, y_train_weather, y_test_weather = train_test_split(X, y_weather_type, test_size=0.2, random_state=42)
```

## Decision Tree

```python
# Mô hình dự đoán khả năng mưa
rain_model_dt = DecisionTreeRegressor(random_state=42)
rain_model_dt.fit(X_train, y_train_rain)
y_pred_rain_dt = rain_model_dt.predict(X_test)

# Mô hình dự đoán loại thời tiết
weather_model_dt = DecisionTreeClassifier(random_state=42, class_weight='balanced')
weather_model_dt.fit(X_train, y_train_weather)
y_pred_weather_dt = weather_model_dt.predict(X_test)
```

```python
# Thông số kết quả của mô hình dự đoán khả năng mưa
print(f'Decision Tree Regressor Result:')
mse_dt = mean_squared_error(y_test_rain, y_pred_rain_dt)
mea_dt = mean_absolute_error(y_test_rain, y_pred_rain_dt)
r2_dt = r2_score(y_test_rain, y_pred_rain_dt)  # Hệ số xác định R^2
print(f'MSE: {round(mse_dt/100, 2):.2f}')
print(f'MEA: {round(mea_dt/100, 2):.2f}')
print(f'R^2 Score: {round(r2_dt, 2):.2f}')
```

```python
# Thông số kết quả của mô hình dự đoán loại thời tiết
print(f'Decision Tree Classifier Result:')
accuracy_dt = accuracy_score(y_test_weather, y_pred_weather_dt)
precision_dt = precision_score(y_test_weather, y_pred_weather_dt, average='weighted', zero_division=0)
recall_dt = recall_score(y_test_weather, y_pred_weather_dt, average='weighted', zero_division=0)
f1_dt = f1_score(y_test_weather, y_pred_weather_dt, average='weighted', zero_division=0)
print(f'Accuracy: {round(accuracy_dt, 2):.2f}')
print(f'Precision: {round(precision_dt, 2):.2f}')
print(f'Recall: {round(recall_dt, 2):.2f}')
print(f'F1 Score: {round(f1_dt, 2):.2f}')
```

## Random Forest

```python
# Mô hình dự đoán khả năng mưa
rain_model_rf = RandomForestRegressor(n_estimators=3, random_state=42)
rain_model_rf.fit(X_train, y_train_rain)
y_pred_rain_rf = rain_model_rf.predict(X_test)

# Mô hình dự đoán loại thời tiết
weather_model_rf = RandomForestClassifier(n_estimators=10, random_state=42, class_weight='balanced')
# weather_model = RandomForestClassifier(n_estimators=7, class_weight={0: 9, 1:2, 2:3, 3:20, 4:2, 5:1, 6:14, 7:3, 8:0, 9:11})
weather_model_rf.fit(X_train, y_train_weather)
y_pred_weather_rf = weather_model_rf.predict(X_test)
```

```python
# Thông số kết quả của mô hình dự đoán khả năng mưa
print(f'Random Forest Regressor Result:')
mse_rf = mean_squared_error(y_test_rain, y_pred_rain_rf)
mea_rf = mean_absolute_error(y_test_rain, y_pred_rain_rf)
r2_rf = r2_score(y_test_rain, y_pred_rain_rf)  # Hệ số xác định R^2
print(f'MSE: {round(mse_rf, 2):.2f}')
print(f'MEA: {round(mea_rf, 2):.2f}')
print(f'R^2 Score: {round(r2_rf, 2):.2f}')
```

```python
# Thông số kết quả của mô hình dự đoán loại thời tiết
print(f'Random Forest Classifier Result:')
accuracy_rf = accuracy_score(y_test_weather, y_pred_weather_rf)
precision_rf = precision_score(y_test_weather, y_pred_weather_rf, average='weighted', zero_division=0)
recall_rf = recall_score(y_test_weather, y_pred_weather_rf, average='weighted', zero_division=0)
f1_rf = f1_score(y_test_weather, y_pred_weather_rf, average='weighted', zero_division=0)
print(f'Accuracy: {round(accuracy_rf, 2):.2f}')
print(f'Precision: {round(precision_rf, 2):.2f}')
print(f'Recall: {round(recall_rf, 2):.2f}')
print(f'F1 Score: {round(f1_rf, 2):.2f}')
```
